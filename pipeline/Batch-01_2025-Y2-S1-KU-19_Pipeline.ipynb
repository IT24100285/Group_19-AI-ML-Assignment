**Outlier Detection & Removal**

```python
def make_numeric_New_Crop_Table_ohe(orig_New_Crop_Table_ohe):
    # convert strings with commas like "1,234" -> 1234
    numeric_New_Crop_Table_ohe = orig_New_Crop_Table_ohe.apply(lambda s: pd.to_numeric(s.astype(str).str.replace(',','').str.strip(), errors='coerce'))
    return numeric_New_Crop_Table_ohe

numeric_New_Crop_Table_ohe = make_numeric_New_Crop_Table_ohe(New_Crop_Table_ohe)
numeric_cols = [c for c in numeric_New_Crop_Table_ohe.columns if numeric_New_Crop_Table_ohe[c].notna().sum() > 0]
print("\nDetected numeric-ish columns (will be used for outlier checks):")
print(numeric_cols)

# show basic stats for numeric columns
if numeric_cols:
    display(numeric_New_Crop_Table_ohe[numeric_cols].describe().T)
else:
    print("No numeric columns detected. Check your CSV or column parsing.")
```

```python
import math
from scipy import stats  # SciPy present in Colab; used for z-score

def iqr_bounds(series, k=1.5):
    # series should be numeric with NaNs allowed
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - k * iqr
    upper = q3 + k * iqr
    return lower, upper

def detect_outliers_iqr(series, k=1.5):
    lower, upper = iqr_bounds(series, k=k)
    return (series < lower) | (series > upper)

def detect_outliers_zscore(series, thresh=3.0):
    if series.dropna().shape[0] == 0:
        return pd.Series(False, index=series.index)
    std = series.std(ddof=0)
    if std == 0 or math.isnan(std):
        # constant series -> no z-score outliers
        return pd.Series(False, index=series.index)
    z = (series - series.mean()) / std
    return z.abs() > thresh

def outlier_summary(New_Crop_Table_ohe_numeric, method='iqr', k=1.5, z_thresh=3.0):
    # returns dict: {col: number_of_outliers}
    out_counts = {}
    for c in New_Crop_Table_ohe_numeric.columns:
        s = New_Crop_Table_ohe_numeric[c]
        if s.dropna().shape[0] == 0:
            out_counts[c] = 0
            continue
        if method == 'iqr':
            mask = detect_outliers_iqr(s, k=k)
        elif method == 'z':
            mask = detect_outliers_zscore(s, thresh=z_thresh)
        else:
            raise ValueError("method must be 'iqr' or 'z'")
        out_counts[c] = int(mask.sum())
    return out_counts
```

```python
if not numeric_cols:
    print("No numeric columns to analyze.")
else:
    iqr_counts = outlier_summary(numeric_New_Crop_Table_ohe[numeric_cols], method='iqr', k=1.5)
    z_counts   = outlier_summary(numeric_New_Crop_Table_ohe[numeric_cols], method='z', z_thresh=3.0)

    summary_New_Crop_Table_ohe = pd.DataFrame({
        'iqr_outliers': pd.Series(iqr_counts),
        'z_outliers': pd.Series(z_counts),
        'non_nulls': numeric_New_Crop_Table_ohe[numeric_cols].notna().sum()
    })
    display(summary_New_Crop_Table_ohe.sort_values('iqr_outliers', ascending=False))
```

```python
cols_to_check = numeric_cols
k = 1.5
how = 'any'

def remove_outliers_iqr(New_Crop_Table_ohe_orig, New_Crop_Table_ohe_numeric, cols=None, k=1.5, how='any'):
    if cols is None:
        cols = New_Crop_Table_ohe_numeric.columns.tolist()
    if how not in ('any','all'):
        raise ValueError("how must be 'any' or 'all'")
    masks = []
    for c in cols:
        s = New_Crop_Table_ohe_numeric[c]
        lower, upper = iqr_bounds(s, k=k)
        mask = (s < lower) | (s > upper)
        masks.append(mask.fillna(False))
    if not masks:
        return New_Crop_Table_ohe_orig.copy(), 0
    if how == 'any':
        combined = pd.concat(masks, axis=1).any(axis=1)
    else:
        combined = pd.concat(masks, axis=1).all(axis=1)
    removed_count = int(combined.sum())
    New_Crop_Table_ohe_cleaned = New_Crop_Table_ohe_orig.loc[~combined].copy()
    return New_Crop_Table_ohe_cleaned, removed_count

New_Crop_Table_ohe_iqr_clean, removed_iqr = remove_outliers_iqr(New_Crop_Table_ohe, numeric_New_Crop_Table_ohe, cols=cols_to_check, k=k, how=how)
print(f"IQR removal ('{how}', k={k}): removed {removed_iqr} rows. New shape: {New_Crop_Table_ohe_iqr_clean.shape} (old: {New_Crop_Table_ohe.shape})")
```
